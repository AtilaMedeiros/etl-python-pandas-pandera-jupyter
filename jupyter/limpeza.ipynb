```python
df.set_index('codigo_ocorrencia', inplace=True) #deixa de enumerar as linhas e usa o codigo_ocorrencia para numerar. inplace= True aplica as alterações
df.head()
```

```python
df.reset_index(drop=True, inplace=True) #reinicia o index(reset_index), remove as alterações(drop-True) e aplica tudo na base de dados(inplace=True)
df.head()
```

```python
df.loc[40] #localize a coluna index o valor 40
```

```python
df.loc[0] = 20 # altera todos os valores da linha 0 para 20
df.head()
```

```python
df.loc[:, "ocorrencia_uf"] = "CE" # altera todos os valores da coluna para CE
df.head()
```

```python
df['ocorrencia_cidade_bkp'] = df.ocorrencia_cidade #faz um backup da coluna ocorrenci_cidade
df.head()
```

```python
df.loc[df.ocorrencia_classificacao == 'INCIDENTE', ['ocorrencia_uf']] = 'SP' #localize na coluna "ocorrencia_classificacao" toda linha que for 'INCIDENTE' e atualize a coluna 'ocorrencia_uf' com valor 'SP'
df.head()
```

```python
df.loc[df.ocorrencia_uf == 'SP'] #filtra mas não aplica todos os valores SP na coluna ocorrencia_uf
df.head()
```

```python
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA #localize todas as linhas com da coluna ocorrencia_aerodromo com valor '****' e altere pelo valor padrão NA do Pandas. NA = sem valor
df.head()
```

```python
df.replace(['**', '###!', '####', '****', '*****', 'NULL'], pd.NA, inplace=True ) # procure os valores da lista ['***', ...] e substitua pelo valor <NA> (sem valor) e aplique as alterações inplace=True
df.tail()
```

```python
df.isna() # retorna true para os valores que foram setados pela função pd.NA. Podemos usar tambem a função df.isnull() que é a mesma coisa 
```

```python
df.isna().sum() #soma a quantidade de dados que não tem atribuição. Quando usamos o df.NA. 
```

```python
df.fillna(0, inplace=True) # preencha dodas as celulas que tem valor <NA> (df.NA) com valor zero e aplique na variavel DF. Não temos alteração do arquivo CSV
```

```python
df.replace([0], pd.NA, inplace=True) #retorna o valor de 0 para não atribuido (df.NA).
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(value={'total_recomendacoes': 10}, inplace=True) #altere todos oa valores definido como df.na, da coluna total_recomendacoes, para 10 e aplica as alterações na var df
df.head()
```

```python
df.loc[df.total_recomendacoes == 10] #filtra todos os dados da coluna total_recomendacores com o valor 10, mas não aplica na var df
```

```python
df['total_recomendacoes_bck'] = df.total_recomendacoes
df
```

```python
df.drop(['total_recomendacoes_bck'], axis=1, inplace=True) #apaga a coluna com nome 'total_recomendacoes'. Por padrão o eixo é zero (axis=0) e o pandas vai tentar excluir uma linha que tenho o nome setado. Precisamos alterar para o eixo 1 (coluna) para apagar a coluna correta. Sem o inplace o panda retira a coluna somente da visualização e não do data frame
df.head()
```

```python
df.dropna() #apaga a linha toda, quando ela tem alguma valor NA
```

```python
df.dropna(subset=['ocorrencia_aerodromo']) #apaga uma linha quando ela tem algum valor NA, na coluna 'ocorrencia_uf'
```

```python
df.drop_duplicates(inplace=True) #apaga as linhas duplicadas. 
df
```
